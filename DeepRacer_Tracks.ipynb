```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

track = np.load("2022_may_open.npy")
```

```python
plt.rcParams['figure.figsize'] = [50, 50]
for x in range(int((track.size)/6)):
  waypoint = track[x]
  str1="$"
  str2=str(x+1)
  str3=str1+str2+str1
  for y in range(3):
    waypointx = waypoint[2*y]
    waypointy = waypoint[2*y+1]
    plt.scatter(waypointx, waypointy, marker=str3, s=500, c='black')

plt.savefig("2022mayopentrack.png")
files.download("2022mayopentrack.png")
```
